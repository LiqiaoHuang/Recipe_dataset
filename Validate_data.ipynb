```python
import os
import glob

def read_nutrition_info(path):
    files = glob.glob(os.path.join(path, '*.csv'))
    data = {}
    for file in files:
        df = pd.read_csv(file)
        ingredient = os.path.basename(file).split('.')[0]
        # 根据列名选择相应的数据
        amount_col = 'Average Amount' if 'Average Amount' in df.columns else 'Amount'
        for _, row in df.iterrows():
            if row['Name'] not in data:
                data[row['Name']] = {}
            data[row['Name']][ingredient] = row[amount_col]
    return pd.DataFrame(data)
```

```python
nutrition_info = read_nutrition_info('nutrition_amount_final')
```

```python
def calculate_nutrition(row, nutrition_info):
    total_nutrition = {}
    for ingredient, weight in row.items():
        if ingredient in nutrition_info.columns:
            nutrition = nutrition_info[ingredient] * weight / 100
            for nutrient, amount in nutrition.items():
                if nutrient not in total_nutrition:
                    total_nutrition[nutrient] = 0
                total_nutrition[nutrient] += amount
    return pd.Series(total_nutrition)

df_nutrition = df.apply(lambda row: calculate_nutrition(row, nutrition_info), axis=1)
```

```python
import pandas as pd

# 读取数据
df1 = pd.read_excel('output.xlsx', index_col=0)
df2 = pd.read_excel('output_nutrient_processed.xlsx', index_col=0)

# 初始化一个空的DataFrame来保存每道菜的所有营养的含量
result_df = pd.DataFrame(index=df1.index, columns=df2.columns)

# 遍历df1的每一行
for idx, row in df1.iterrows():
    print(f"Processing dish: {idx}")

    # 初始化一个空的Series来保存这道菜的所有营养的含量
    nutrition_series = pd.Series(index=df2.columns).fillna(0)  # 将NaN值初始化为0
    
    # 遍历df1的每一个原材料列，包括重复的列
    for ingredient in df1.columns:
        weight = row[ingredient]
        print(f"Ingredient: {ingredient}, Weight: {weight}")
        
        # 如果这种原材料在df2中
        if ingredient in df2.index and not pd.isna(weight):
            print("Found ingredient in df2")
            nutrition_series += df2.loc[ingredient] * weight
        else:
            print("Ingredient not found in df2 or weight is NaN")

    # 将这道菜的所有营养的含量保存到结果DataFrame中
    result_df.loc[idx] = nutrition_series
result_df = result_df.div(100)
result_df
```

```python
result_df.to_excel("output_results.xlsx")
```
